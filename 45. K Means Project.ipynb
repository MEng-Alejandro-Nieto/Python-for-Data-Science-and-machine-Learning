```python
from sklearn.cluster import KMeans
```

```python
dummi=pd.get_dummies(data_result['Private'],drop_first=True)
data_result=data_result.join(dummi).head(3)
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(data_result.drop('Private',axis=1))
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(data_result['Yes'],kmeans.labels_)
```
